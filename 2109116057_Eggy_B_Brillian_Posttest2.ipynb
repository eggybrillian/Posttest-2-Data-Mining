```python
import pandas as pd
import seaborn  as sns
```

```python
df = pd.read_csv("tabulasi.csv")
df
```

```python
df.sum()
```

```python
print("List Kolom :\n")
for col in df.columns:
  print(col)
```

```python
from itertools import permutations #Library untuk buat kombinasi barang
combi_all = list(permutations(df.columns, 2)) # untuk menyimpan seluruh kemungkinan kombinasi barang
combi_all # untuk menampilkan keseluruhan hasil kombinasi
```

```python
hasil = [] # Menyimpan data dalam bentuk list

for combi_item in combi_all: # Mengulangi pembacaan dataset
  if len(df[(df[combi_item[0]]==1) & (df[combi_item[1]]==1)]) >0: # Menentukan apakah kombinasi tersebut ada atau tidak
    hasil.append(combi_item) # Menyimpan hasil pengecekan dalam bentuk tuple didalam list

hasil # Menampilkan hasil
```

```python
hasil_akhir = {} # Menyimpan data dalam bentuk dictionary

for itemset in hasil:  # Mengulangi pembacaan dataset
  banyak_data= len(df[(df[itemset[0]]==1) & (df[itemset[1]]==1)]) # Menentukan berapa jumlah kemunculan kombinasi dan menyimpannya dalam sebuah variabel
  hasil_akhir[str(itemset)] = banyak_data # Menambahkan hasil penentuan ke dalam dictionary
  
hasil_akhir # Menampilkan hasil
```

```python
for item,ab in hasil_akhir.items(): # Untuk melakukan perulangan pembacaan dataset pada tiap - tiap kombinasi itemnya
    support = ab/len(df)*100 # Menghitung support pada tiap tiap kombinasi item
    print(f"{item} support = {support:.3f}%") # Menampilkan  nilai support
```

```python
for item,ab in hasil_akhir.items(): # Untuk melakukan perulangan pembacaan dataset pada tiap - tiap kombinasi itemnya
  i = eval(item) # Ubah ke list dari string
  confidence = ab/df[i[0]].sum()*100 # Menghitung confidence pada tiap tiap kombinasi item dan menampilkannya
  print(f"{i[0]}->{i[1]}  Confidence = {confidence:.1f}%") # Menampilkan hasil confidence
```

```python
df2 = pd.read_csv('deteksi_uang_palsu.csv')
df2
```

```python
# 1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi

data_prediksi = [172.25, 103.94, 103.68, 4.14, 3.25, 112.48]

# tambahkan kolom jarak
df2['distance'] = ((df2['diagonal'] - data_prediksi[0])**2 
               + (df2['height_left'] - data_prediksi[1])**2
               + (df2['height_right'] - data_prediksi[2])**2
               + (df2['margin_low'] - data_prediksi[3])**2
               + (df2['margin_up'] - data_prediksi[4])**2
               + (df2['length'] - data_prediksi[5])**2)**0.5

# Menampilkan dataset yang sudah ditambahkan dengan kolom jarak
print(df2)
```

```python
# 2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)

import math

# Mengurutkan data berdasarkan jarak terdekat
k = int(math.sqrt(len(df2)))
sorted_df2 = df2.sort_values(by=["distance"]).head(k)

# Menampilkan hasil pengurutan
sorted_df2
```

```python
# 3. Tentukan apa hasil prediksinya berdasarkan label terbanyak

hasil_prediksi = sorted_df2["is_genuine"].value_counts().idxmax()
print("Hasil prediksi: ", hasil_prediksi)
```

#Data Preparation :
Tahap persiapan data :

*   Menghapus variabel yang tidak relevan atau duplikat seperti customer_id karena tidak digunakan dalam analisis.
*   Mengubah variabel kategorikal menjadi variabel dummy, seperti gender dan country.
*   Melakukan imputasi atau pengisian nilai kosong pada variabel balance jika ada, dengan menggunakan nilai rata-rata atau median.
*   Melakukan scaling pada variabel numerik seperti age, tenure, balance, products_number, credit_score, dan estimated_salary, agar variabel memiliki skala yang serupa dan dapat dibandingkan secara seimbang.
*   Memastikan tidak ada nilai yang tidak valid atau outliers dalam dataset.

```python
#Menghapus kolom customer_id dan kemmudian menampilkan dataset nya
df3 = df3.drop(['customer_id'], axis=1)
df3
```

```python
#Mengubah variabel kategorikal pada gender dan country menjadi variabel dummy
#parameter drop_first=True digunakan untuk menghindari jebakan dummy variabel, 
#yaitu agar variabel dummy pertama tidak digunakan dalam analisis.
df3 = pd.get_dummies(df3, columns=['gender', 'country'], drop_first=True)

#menampilkan dataset
df3
```

```python
if df3['balance'].isnull().values.any(): # Cek apakah ada nilai kosong pada variabel Balance
    balance_mean = df3['balance'].mean() # Hitung nilai rata-rata dari variabel Balance
    df3['balance'].fillna(balance_mean, inplace=True)  # Isi nilai kosong pada variabel Balance dengan nilai rata-rata

# Cek apakah ada nilai kosong pada variabel Balance setelah imputasi
if df3['balance'].isnull().values.any():
    print("Jumlah nilai kosong pada variabel balance setelah imputasi:", df3['balance'].isnull().sum())
else:
    print("Tidak ada nilai kosong pada variabel balance setelah imputasi.")  
```

```python
from sklearn.preprocessing import StandardScaler

# Mengambil variabel numerik
numeric_vars = ['age', 'tenure', 'balance', 'products_number', 'credit_score', 'estimated_salary']

# Melakukan Standard Scaling pada variabel numerik
scaler = StandardScaler()
df3[numeric_vars] = scaler.fit_transform(df3[numeric_vars])

# Menampilkan dataset
df3
```

```python
# Mengambil variabel numerik
numeric_vars = ['age', 'tenure', 'balance', 'products_number', 'credit_score', 'estimated_salary']

# Mencari nilai outliers dengan Z-score
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df3[numeric_vars]))
threshold = 3
outliers = np.where(z_scores > threshold)

# Menampilkan nilai outliers
print(outliers)
```
